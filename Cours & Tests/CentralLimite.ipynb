```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import exp, sqrt, pi
```

```python
domaine = range(-100, 100)
mu = [-10,0,35,80]
sigma = [10,20,30,50]
f = lambda x, mu, sigma : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))
y = [[f(x,mu[i], sigma[i]) for x in domaine] for i in range(len(mu))]
figure = plt.figure()
for i in range(len(mu)):
    plot = plt.plot(domaine, y[i])
```
