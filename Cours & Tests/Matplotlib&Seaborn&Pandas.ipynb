# Matplotlib Tests

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,1000)
```

```python
plt.rcParams.update({'font.size' : 10})
ax = plt.axes()
ax.set(xlabel='x', ylabel='sin(x)')
plt.axis([0, 10, -1, 1])
plt.title('Mon premier graph')
plt.axis([0, 10, -1, 1])
plt.title('Mon premier graph')
plt.plot(x, np.sin(x), color = 'black', linestyle='dotted', label='sin(x)')
plt.legend(loc='lower left')
```

```python
x1 = np.linspace(0, 10, 50)
```

```python
dy = 0.2
```

```python
figure = plt.figure()
y1 = np.sin(x1) + dy * np.random.randn(50)
y2 = np.sin(x1) + dy
plt.errorbar(x1, y1, yerr=dy, fmt='ok', ecolor='lightgray', elinewidth=3, capsize=5, capthick=2, barsabove=False)
plt.plot(x1,y1)
plt.plot(x1,y2)
```

```python
# Une fonction pour gnrer facilement des DataFrame. 
# Elle nous sera utile dans la suite de ce chapitre.
def make_df(cols, ind):
    """Cre rapidement des DataFrame"""
    data = {c: [c + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)
```

```python
# exemple
make_df('ABC', range(3))
```

```python
df1 = make_df('AB', [1,2])
df1
```

```python
df2 = make_df('AB', [3,4])
df2
```

## JOIN: merge

```python
# NATURAL JOIN
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'department': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'date': [2004, 2008, 2012, 2014]})
# 1 to 1                    
df3 = pd.merge(df1, df2)
df3
```

```python
# JOIN ON
# df4 = pd.merge(df1, df2, left_on= "employee", right_on= "emp_name")
```

```python
df4 = pd.DataFrame({'department': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})
# 1 to *
pd.merge(df3, df4)
```

```python
df5 = pd.DataFrame({'department': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'competence': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})
# * to *
pd.merge(df1, df5)
```
