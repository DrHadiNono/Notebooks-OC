```python
from collections import Counter

def most_common_words(labels):
    words = []
    for lab in labels:
        words += [word for word in lab.split(' ') if len(word)>2 or word=='CB']
    counter = Counter(words)
    for word in counter.most_common(1000):
        print(word)

most_common_words(operations_data['libelle'].values)
```

```python
CATEGS = {
    'LOYER': 'LOYER',
    'ETRANGER': 'DON',
    'CARTE': 'COTISATION BANCAIRE',   
    'ASSURANCE': 'COTISATION BANCAIRE',
    'TENUE DE COMPTE': 'COTISATION BANCAIRE',
    'SERVICES': 'COTISATION BANCAIRE',
    'OFFRE': 'COTISATION BANCAIRE',
    'MGEN': 'SANTE',
    'DR' : 'SANTE',
    'PH.' : 'SANTE',
    'AMAZON': 'ACHAT & MEDIA',
    'PAYPAL': 'ACHAT & MEDIA',
    'BLADE': 'ACHAT & MEDIA',
    'COFIDIS': 'ACHAT & MEDIA',
    'DECATHLON': 'SPORT',
    'LIFESTYLE': 'SPORT',
    'GEOX': 'VETEMENTS',
    'INTER ': 'COURSES',
    'IMTIYAZ': 'COURSES',
    'CARREFOUR': 'COURSES',
    'AUCHAN': 'COURSES',
    'LIDL': 'COURSES',
    'LAV': 'LAVERIE',
    'ORANGE': 'FACTURE TELEPHONE',
    'CORIOLIS': 'FACTURE TELEPHONE',    
    'TRANSP': 'TRANSPORT',
    'SNCF': 'TRANSPORT',
    'RATP': 'TRANSPORT',
    'HOTEL' : 'HOTEL',
    'IBIS' : 'HOTEL',
    'UBER *EATS' : 'FOOD',
    'TACOS' : 'FOOD',
    'SUBWAY' : 'FOOD',
    'DONALD' : 'FOOD',
    'SUSHI' : 'FOOD',
    'CARLIER' : 'FOOD',
    'PITAYA' : 'FOOD',
    'PIZZA' : 'FOOD',
}

TYPES = {
    'CB': 'CARTE',
    'VIR ': 'VIREMENT',
    'VIREMENT': 'VIREMENT',
    'RETRAIT': 'RETRAIT',
    'PRLV': 'PRELEVEMENT',  
}
```

```python
EXPENSES = [50, 150] # Bornes des catégories de dépense : petite, moyenne et grosse
LAST_BALANCE = 301.18 # Solde du compte APRES la dernière opération en date
WEEKEND = ["Saturday","Sunday"] # Jours non travaillés
```

```python
# Controle des colonnes
for c in ['date_operation','libelle','debit','credit']:
    if c not in operations_data.columns:
        if (c in ['debit','credit'] and 'montant' not in operations_data.columns) or \
        (c not in ['debit','credit']):
            msg = "Il vous manque la colonne '{}'. Attention aux majuscules "
            msg += "et minuscules dans le nom des colonnes!"
            raise Exception(msg.format(c))

# Suppression des colonnes innutiles
for c in operations_data.columns:
    if c not in ['date_operation','libelle','debit','credit','montant']:
        del operations_data[c]

# Ajout de la colonne 'montant' si besoin
if 'montant' not in operations_data.columns:
    operations_data["debit"] = operations_data["debit"].fillna(0)
    operations_data["credit"] = operations_data["credit"].fillna(0)
    operations_data["montant"] = operations_data["debit"] + operations_data["credit"]
    del operations_data["credit"], operations_data["debit"]

# creation de la variable 'solde_avt_ope'
operations_data = operations_data.sort_values("date_operation")
amount = operations_data["montant"]
balance = amount.cumsum()
balance = list(balance.values)
last_val = balance[-1]
balance = [0] + balance[:-1]
balance = balance - last_val + LAST_BALANCE
operations_data["solde_avt_ope"] = balance

# Assignation des operations_data a une categorie et a un type
def detect_words(values, dictionary):
    result = []
    for lib in values:
        operation_type = "AUTRE"
        for word, val in dictionary.items():
            if word in lib:
                operation_type = val
        result.append(operation_type)
    return result
operations_data["categ"] = detect_words(operations_data["libelle"], CATEGS)
operations_data["type"] = detect_words(operations_data["libelle"], TYPES)

# creation des variables 'tranche_depense' et 'sens'
def expense_slice(value):
    value = -value # Les dépenses sont des nombres négatifs
    if value < 0:
        return "(pas une depense)"
    elif value < EXPENSES[0]:
        return "petite"
    elif value < EXPENSES[1]:
        return "moyenne"
    else:
        return "grosse"
operations_data["tranche_depense"] = operations_data["montant"].map(expense_slice)
operations_data["sens"] = ["credit" if m > 0 else "debit" for m in operations_data["montant"]]

# Creation des autres variables
operations_data["annee"] = operations_data["date_operation"].map(lambda d: d.year)
operations_data["mois"] = operations_data["date_operation"].map(lambda d: d.month)
operations_data["jour"] = operations_data["date_operation"].map(lambda d: d.day)
operations_data["jour_sem"] = operations_data["date_operation"].map(lambda d: d.day_name())
operations_data["jour_sem_num"] = operations_data["date_operation"].map(lambda d: d.weekday()+1)
operations_data["weekend"] = operations_data["jour_sem"].isin(WEEKEND)
operations_data["quart_mois"] = [int((jour-1)*4/31)+1 for jour in operations_data["jour"]]
        
# Enregistrement au format CSV
operations_data.to_csv("operations_data_enrichies.csv",index=False)
```
